```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("data_set.csv") 
```

```python
# Scatter Plot: Product Price vs. Original Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['product_original_price'], y=data['product_price'])
plt.title('Product Price vs. Original Price')
plt.xlabel('Product Original Price')
plt.ylabel('Product Price')
plt.show()
```

```python
# Scatter Plot: Number of Ratings vs. Star Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['product_num_ratings'], y=data['product_star_rating'])
plt.title('Product Number of Ratings vs. Star Rating')
plt.xlabel('Product Number of Ratings')
plt.ylabel('Product Star Rating')
plt.show()
```
